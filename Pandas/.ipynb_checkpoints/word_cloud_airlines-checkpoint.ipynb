```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load in the dataframe
df = pd.read_csv('../Resources/Airline-Sentiment-2-w-AA.csv',  encoding = 'ISO-8859-1')
```

```python
df.head()
```

```python
print("There are {} observations and {} features in this dataset. \n".format(df.shape[0],df.shape[1]))

```

```python
df[["airline", "text","tweet_created"]].head()
```

```python
# Groupby by aireline
airline = df.groupby("airline")

# Summary statistic of all airline
airline.describe().head()
```

```python
airline.mean().sort_values(by="airline_sentiment:confidence",ascending=False).head()
```

```python
plt.figure(figsize=(15,10))
airline.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Airline")
plt.ylabel("Airline Sentiment")
plt.show()
```

```python
#Start with one review:
text = df.text[0]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
text = " ".join(review for review in df.text)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["VirginAmerica", "American", "Delta", "Southwest", "United", "US Airways"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
